```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
test1 = pd.read_csv('fold_1_predictions_test.csv')
test2 = pd.read_csv('fold_2_predictions_test.csv')
test3 = pd.read_csv('fold_3_predictions_test.csv')
test4 = pd.read_csv('fold_4_predictions_test.csv')
test5 = pd.read_csv('fold_5_predictions_test.csv')
test6 = pd.read_csv('fold_6_predictions_test.csv')
test7 = pd.read_csv('fold_7_predictions_test.csv')
test8 = pd.read_csv('fold_8_predictions_test.csv')
test9 = pd.read_csv('fold_9_predictions_test.csv')
test10 = pd.read_csv('fold_10_predictions_test.csv')
```

```python
test = pd.concat([test1,test2,test3,test4,test5,test6,test7,test8,test9,test10],axis=0)
test.to_csv('test.csv')
print(test.shape)
```

```python
pre = test['Predicted Value']
EXP = test['Experimental Value']
print('MAPE:{} R2:{} MAE:{} RMSE:{}'.format(mean_absolute_percentage_error(EXP,pre),r2_score(EXP,pre),mean_absolute_error(EXP,pre),np.sqrt(mean_squared_error(EXP,pre))))
```

```python
r2 = []
MAPES = []
MAES = []
for i in range(1, 10):
    df = pd.read_csv('fold_{}_predictions_test.csv'.format(i))
    pre = df['Predicted Value']
    EXP = df['Experimental Value']
    r2_ = r2_score(EXP, pre)
    MAPE = mean_absolute_percentage_error(EXP, pre)
    MAE = mean_absolute_error(EXP, pre)
    MAES.append(MAE)
    MAPES.append(MAPE)
    r2.append(r2_)
print(np.mean(r2), np.std(r2))
print(np.mean(MAPES), np.std(MAPES))
print(np.mean(MAES), np.std(MAES))
```
